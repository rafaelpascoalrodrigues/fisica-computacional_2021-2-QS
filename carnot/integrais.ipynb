```python
import numpy as np
import matplotlib.pyplot as plt

def integracao_trapezios(f, a: float = 0.0, b: float = 1.0, n: int = 10)->float:
    h = (b - a) / n
    s = 0.5 * f(a) + 0.5 * f(b)
    for k in range(1, n):
        s += f(a + k * h)

    return h * s

def isotherm(n = 1, R = 0.082, T = 400, V = 10): return n * R * T / V
def isotherm_hot(x): return isotherm(V = x, T = 400)
def isotherm_cold(x): return isotherm(V = x, T = 200)

xvals, plot_hot, plot_cold = [], [], []
for x in np.arange(4.0, 9.1, 0.1):
    xvals += [x]
    plot_hot += [isotherm_hot(x)]
    plot_cold += [isotherm_cold(x)]

fig, axs = plt.subplots(1, 1, figsize=(24, 4), dpi = 180)
axs.set_xlabel(r'V(m^3)')
axs.set_ylabel(r'$J_m$')
axs.plot(xvals, plot_hot, 'rx-', label=r"$Isoterma (400 K)$")
axs.plot(xvals, plot_cold, 'bx-', label=r"$Isoterma (200 K)$")
axs.legend()
    
print("Área aproximada abaixo da Isoterma 200 K:", integracao_trapezios(isotherm_cold, 4, 9, 20))
print("Área aproximada abaixo da Isoterma 400 K:", integracao_trapezios(isotherm_hot, 4, 9, 20))
```
